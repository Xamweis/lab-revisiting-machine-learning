```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

def predict_col(numericals:pd.DataFrame, column, features):
    model_data = numericals.dropna(subset=[column])
    X = pd.DataFrame(model_data[features])
    y = model_data[column]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
    transformer = MinMaxScaler().fit(X_train)
    X_train_norm = transformer.transform(X_train)
    X_test_norm = transformer.transform(X_test)

    model = LogisticRegression()
    model.fit(X_train_norm, y_train)

    nans = numericals[numericals[column].isna()]
    not_nans = numericals.dropna(subset=[column])

    nans[column] = model.predict(
        transformer.transform(
            nans[features]))

    print('R2:', r2_score(y_test, model.predict(X_test_norm)))
    numericals = pd.concat([nans, not_nans], axis=0)
    return numericals
```

```python
data = pd.read_csv('data/treated.csv')
data
```

```python
# from tha lessons:
drop_list = [
    'MDMAUD', 'NOEXCH','MAILCODE', 'MDMAUD_R', 'MDMAUD_F','MDMAUD_A'
    ]

for col_name in data.columns:
    if "RFA" in col_name or "ADATE" in col_name:
        drop_list.append(col_name)

drop_list.remove('RFA_2R')
drop_list.remove('RFA_2A')

data = data.drop(drop_list, axis=1)

data['CLUSTER'] = data['CLUSTER'].fillna('40')
data['HOMEOWNR'] = data['HOMEOWNR'].fillna('U')
data['DATASRCE'] = data['DATASRCE'].fillna('3')
data['INCOME'] = data['INCOME'].fillna(5.0)
data['CLUSTER2'] = data['CLUSTER2'].fillna(np.ceil(np.mean(data['CLUSTER2'])))
data['TIMELAG'] = data['TIMELAG'].fillna(np.median(data['TIMELAG']))
```

```python
data['WEALTH1'] = data['WEALTH1'].astype(object)
data['WEALTH2'] = data['WEALTH2'].astype(object)
data['CLUSTER'] = data['CLUSTER'].astype(object)
data['INCOME'] = data['INCOME'].astype(object)
data['DATASRCE'] = data['DATASRCE'].astype(object)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

X = nums_scaled
y = targets['TARGET_B']

kbest = SelectKBest(chi2, k=10).fit(X, y)

df = pd.DataFrame(data = kbest.scores_, columns = ['score'])
df['Column'] = nums.columns
display(df.sort_values(by = ['score'], ascending = False).head(20))
```
